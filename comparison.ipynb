```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# the dataset

# Test dataset
# test_df_svm = pd.read_csv('svm/test_with_svm_anomalies.csv')
# test_df_iso_f = pd.read_csv('isolation-forest/test_with_final_anomalies.csv')

# Predict dataset
test_df_svm = pd.read_csv('svm/predict_with_svm_anomalies.csv')
test_df_iso_f = pd.read_csv('isolation-forest/predict_with_final_anomalies.csv')
```

```python
# Error count
svm_error = test_df_svm['is_anomalous_svm'].sum()
iso_f_error = test_df_iso_f['is_anomalous'].sum()

print("Isolation Forest Anomalies:", iso_f_error )
print("One-Class SVM Anomalies:", svm_error)
```

```python
# Compare anomaly counts

plt.bar(["Isolation Forest", "One-Class SVM"], [iso_f_error, svm_error], color=['blue', 'green'])
plt.title("Model Comparison: Anomalies Detected")
plt.ylabel("Number of Anomalies")
plt.show()
```
